```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
m = len(x)

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
m = numerator / denominator
c = y_mean - m * x_mean

print(f"Slope (m): {m:.4f}")
print(f"Intercept (c): {c:.4f}")
print(f"Fitted line: y = {m:.4f}x + {c:.4f}")

y_fit = m * x + c

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Original Data')
plt.plot(x, y_fit, color='red', label='Fitted Line: y = {:.2f}x + {:.2f}'.format(m, c))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([0.6, 2.4, 3.5, 4.8, 5.7])
m = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x*y)
sum_x2 = np.sum(x**2)

x_mean = sum_x / m
y_mean = sum_y / m

a1 = ((m*sum_xy) - (sum_x*sum_y)) / ((m*sum_x2) - (sum_x**2))
a0 = (y_mean - (a1*x_mean))

print(f"Slope (a1): {a1:.2f}")
print(f"Intercept (a0): {a0:.2f}")

y_pred = a0 + a1*x

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x, y_pred, color='blue', label='Best_fit line')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Fit')
plt.legend()
plt.grid(True)
plt.show()
```
